```python
import pandas as pd
import requests
import json

def getJson(point1, point2):
    try:
        print(point1 + point2)
        url = 'https://restapi.amap.com/v3/geocode/geo?address=' + point1 + point2 + '&key='
        res = requests.get(url)
        if res.status_code == 200:
            json_text = res.json()
            location = json_text['geocodes'][0]['location'].split(',')
            return location
    except Exception:
        print('error!')
        return None


def S2F(S):
    lng = float(S[0])
    lat = float(S[1])
    height = 0
    return [lng, lat, height]


data_original = pd.read_csv('ruhua_database_details.csv', header=0, encoding="utf-8-sig")
data_original = data_original.dropna(axis=0, how='any')

data_original['position'] = data_original.apply(lambda x: getJson(x['province'], x['local']), axis=1)
data_original['position'] = data_original.apply(lambda x: S2F(x['position']), axis=1)
data_original_json = data_original.to_dict('records')
with open('ruhua.json', 'w', encoding='utf-8') as f:
    json.dump(data_original_json, f, ensure_ascii=False)
```
